```python
import os
import csv
import time
from typing import Literal

from openai import RateLimitError
from IPython.display import Image, display
from langchain.tools import tool
from slack_sdk import WebClient
from slack_sdk.errors import SlackApiError
from tavily import TavilyClient

from deepagents import create_deep_agent
from deepagents.backends import LocalShellBackend

# --------------------------------------------------
# 1. Set environment variables
# Replace these placeholders with your real values
# --------------------------------------------------
os.environ["OPENAI_API_KEY"] = ""
os.environ["TAVILY_API_KEY"] = ""
os.environ["LANGSMITH_API_KEY"] = ""
os.environ["LANGSMITH_TRACING"] = "true"

os.environ["SLACK_BOT_TOKEN"] = ""
os.environ["SLACK_CHANNEL_ID"] = ""

# --------------------------------------------------
# 2. Create sample sales CSV locally in Colab
# --------------------------------------------------
os.makedirs("/content/data", exist_ok=True)

data = [
    ["Date", "Product", "Units Sold", "Revenue"],
    ["2025-08-01", "Widget A", 10, 250],
    ["2025-08-02", "Widget B", 5, 125],
    ["2025-08-03", "Widget A", 7, 175],
    ["2025-08-04", "Widget C", 3, 90],
    ["2025-08-05", "Widget B", 8, 200],
]

csv_path = "/content/data/sales_data.csv"
with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(data)

print("Created CSV:", csv_path)
print("CSV exists:", os.path.exists(csv_path))
print("Files in /content/data:", os.listdir("/content/data"))

# --------------------------------------------------
# 3. Create backend rooted at /content
# --------------------------------------------------
backend = LocalShellBackend(
    root_dir="/content",
    env={
        "PATH": "/usr/bin:/bin",
        "OPENAI_API_KEY": os.environ["OPENAI_API_KEY"],
        "TAVILY_API_KEY": os.environ["TAVILY_API_KEY"],
        "LANGSMITH_API_KEY": os.environ["LANGSMITH_API_KEY"],
        "LANGSMITH_TRACING": os.environ["LANGSMITH_TRACING"],
        "SLACK_BOT_TOKEN": os.environ["SLACK_BOT_TOKEN"],
        "SLACK_CHANNEL_ID": os.environ["SLACK_CHANNEL_ID"],
    },
    virtual_mode=True,
)

# --------------------------------------------------
# 4. Define tools
# --------------------------------------------------
tavily_client = TavilyClient(api_key=os.environ["TAVILY_API_KEY"])
slack_client = WebClient(token=os.environ["SLACK_BOT_TOKEN"])
channel = os.environ["SLACK_CHANNEL_ID"]

@tool(parse_docstring=True)
def internet_search(
    query: str,
    max_results: int = 5,
    topic: Literal["general", "news", "finance"] = "general",
    include_raw_content: bool = False,
) -> dict:
    """Run a web search.

    Args:
        query: search query
        max_results: maximum number of results
        topic: search topic
        include_raw_content: whether to include raw content
    """
    return tavily_client.search(
        query=query,
        max_results=max_results,
        topic=topic,
        include_raw_content=include_raw_content,
    )

@tool(parse_docstring=True)
def slack_send_message(text: str, file_path: str | None = None) -> str:
    """Send message, optionally including an attachment.

    Args:
        text: text content of the message
        file_path: local file path of attachment in the filesystem
    """
    try:
        if not file_path:
            slack_client.chat_postMessage(channel=channel, text=text)
            return "Message sent."

        if not os.path.exists(file_path):
            return f"Slack send failed: file does not exist at {file_path}"

        slack_client.files_upload_v2(
            channel=channel,
            file=file_path,
            filename=os.path.basename(file_path),
            initial_comment=text,
        )
        return "Message with file sent."
    except SlackApiError as e:
        return f"Slack send failed: {e.response.get('error', 'unknown_error')}"

# --------------------------------------------------
# 5. Create the agent
# --------------------------------------------------
agent = create_deep_agent(
    model="gpt-4.1",
    tools=[internet_search, slack_send_message],
    backend=backend,
)

config = {"recursion_limit": 50}

# --------------------------------------------------
# 6. Helpers for running and extracting output
# --------------------------------------------------
def run_agent_collect_messages(agent, input_message, config, max_retries=4, wait_seconds=8):
    collected = []

    for attempt in range(1, max_retries + 1):
        try:
            print(f"Starting attempt {attempt}/{max_retries}...")
            for step in agent.stream(
                {"messages": [input_message]},
                config,
                stream_mode="updates",
            ):
                for _, update in step.items():
                    if update and (messages := update.get("messages")) and isinstance(messages, list):
                        for message in messages:
                            collected.append(message)
                            message.pretty_print()
            return collected
        except RateLimitError:
            if attempt == max_retries:
                raise
            print(f"Rate limited. Waiting {wait_seconds}s before retrying...")
            time.sleep(wait_seconds)

def extract_last_text_message(messages):
    for message in reversed(messages):
        content = getattr(message, "content", None)

        if isinstance(content, str) and content.strip():
            return content

        if isinstance(content, list):
            text_parts = []
            for item in content:
                if isinstance(item, dict) and item.get("type") == "text":
                    text_parts.append(item.get("text", ""))
            if text_parts:
                return "\n".join(text_parts).strip()

    return "No summary text found."

# --------------------------------------------------
# 7. Human-in-the-loop agent request
# --------------------------------------------------
input_message = {
    "role": "user",
    "content": (
        "Analyze /content/data/sales_data.csv. "
        "Create one clean, polished sales visualization and a short summary. "
        "Save the chart locally at /content/sales_plot.png. "
        "Do not send anything to Slack yet. "
        "Return the final summary for human review."
    ),
}

messages = run_agent_collect_messages(agent, input_message, config)

# --------------------------------------------------
# 8. Review summary and plot
# --------------------------------------------------
summary_text = extract_last_text_message(messages)

print("\nFinal summary:\n")
print(summary_text)

plot_path = "/content/sales_plot.png"
print("\nPlot exists:", os.path.exists(plot_path))

if os.path.exists(plot_path):
    display(Image(filename=plot_path))
else:
    print(f"Plot not found at {plot_path}")

# --------------------------------------------------
# 9. Human approval before Slack send
# --------------------------------------------------
approval = input("Send the summary and plot to Slack? (yes/no): ").strip().lower()

if approval == "yes":
    result = slack_send_message.invoke({
        "text": summary_text,
        "file_path": plot_path,
    })
    print(result)
else:
    print("Slack send skipped by human reviewer.")
```
